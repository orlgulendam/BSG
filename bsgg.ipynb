```python
import hashlib
import itertools
import string
import time

# --- AYARLAR ---
TARGET_HASH = "7f85181bc2046bc09096b4988ee01cca"  # Hedef Şifre
MAX_LENGTH = 6  # Kaç haneye kadar tarasın? (Colab'da 5-6 hane idealdir)
CHARSET = string.ascii_lowercase # Sadece küçük harfler (a-z).
# Zorlaştırmak istersen şunu kullan: string.ascii_lowercase + string.digits

def advanced_brute_force():
    print(f"[*] HEDEF HASH: {TARGET_HASH}")
    print(f"[*] KARAKTER SETİ: {CHARSET}")
    print(f"[*] MAKSİMUM UZUNLUK: {MAX_LENGTH}")
    print("-" * 50)

    start_time = time.time()
    total_attempts = 0

    for length in range(1, MAX_LENGTH + 1):
        print(f"\n[+] {length} Haneli Kombinasyonlar Taranıyor...")

        # itertools.product tüm olasılıkları üretir (aa, ab, ac...)
        for attempt_tuple in itertools.product(CHARSET, repeat=length):
            total_attempts += 1

            # Tuple'ı string'e çevir (örn: ('a', 'b') -> "ab")
            attempt_word = "".join(attempt_tuple)

            # MD5 Hash al
            attempt_hash = hashlib.md5(attempt_word.encode('utf-8')).hexdigest()

            # Kontrol et
            if attempt_hash == TARGET_HASH:
                end_time = time.time()
                duration = end_time - start_time

                print("\n" + "=" * 50)
                print(f" >> ŞİFRE KIRILDI! <<")
                print("=" * 50)
                print(f" [V] ŞİFRELİ METİN: {TARGET_HASH}")
                print(f" [V] ÇÖZÜLMÜŞ HALİ: {attempt_word}")
                print("-" * 50)
                print(f" [i] Geçen Süre: {duration:.4f} saniye")
                print(f" [i] Toplam Deneme: {total_attempts} adet")
                print(f" [i] Hız: {total_attempts / duration:.0f} deneme/saniye")
                return # Bulunca dur

            # Her 1 milyon denemede bir bilgi ver (Canlı takip için)
            if total_attempts % 1000000 == 0:
                print(f"    -> {total_attempts} kombinasyon denendi... (Şu anki: {attempt_word})")

    print("\n[-] Belirlenen uzunlukta şifre bulunamadı.")
    print(f"[-] Toplam {total_attempts} deneme yapıldı.")

# Saldırıyı Başlat
advanced_brute_force()
```
